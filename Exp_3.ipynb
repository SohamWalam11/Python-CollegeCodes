```python
# LIST BASICS
# declaring a list
list=[10,10,20,20]

# to print items
print(list[1])
# prints indexed 1st item of the list.

# negative indexing
print(list[-2])

# Acessing elements of Multi-Dimen set
list_1 = [[0,1,2],'Soham',[4,5,6]]
list_2 = [10,20,30]

# print indexed 0 list
print(list_1[0])

# print 4
print(list_1[2][0])

# Methods of List 
''' 1)Append
    2)Insert
    3) Extend'''

# 1)
list_1.append("Hello")
print(list_1)

# 2) 
list_1.insert(0,'crazy')
print(list_1)

# 3) 
list_1.extend(list_2)
print(list_1)

# INPUT METHOD:

car = int(input("ENter Number of range: "))
for i in range(car):
    x = input("ENter something: ")
    list_2.append(x)
print(list_2)

# insert
list_2.insert(0,'Good')
print(list_2)

# pop
list_2.pop()
print(list_2)

# delete
del list_2[2]
print(list_2)

# remove
list_2.remove('Good')
print(list_2)
```
